## Task 1: Linear Approximation
<p>Implement a function that takes a function and a value, and returns a linear approximation of that function at that value.</p>

```python
def plot(m):
    font = {'weight': 'bold', 'size': 9}
    matplotlib.rc('font', **font)
    fig, ax = plt.subplots()
    ax.spines["left"].set_position(("data", 0))
    ax.spines["bottom"].set_position(("data", 0))
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.plot(1, 0, ">k", transform = ax.get_yaxis_transform(), clip_on = False)
    ax.plot(0, 1, "^k", transform = ax.get_xaxis_transform(), clip_on = False)
    x = np.linspace(-30, 30, 101)
    def f(x):
        return np.cbrt(x)
    
    def approx_line(ax: int):
        if ax > 0:
            targets = [i for i in range(1, ax*2+1)]
        else:
            targets = [i for i in range(ax*2, 1)]
        result = [np.cbrt(num) for num in targets]
        potential = []

        for num in result:
            if (num - int(num) <= 0.0000000000001):
                potential.append(np.power(int(num), 3))
        pos = ax; neg = ax
        closest = 0
        while True:
            def checker(pos: int, neg: int):
                if pos in potential:
                    return pos
                elif neg in potential:
                    return neg
                else:
                    return False
            if checker(pos, neg) is False:
                pos += 1
                neg -= 1
            else: 
                closest = checker(pos, neg)
                break

        return(closest)

    def slope(x):
        return 1/(3*np.cbrt(np.power(x, 2)))
    x1 = m
    y1 = f(x1)
    def line(x, x1):
        return slope(x1) * (x - x1) + y1
    ap_y = (approx_line(x1))
    
    x_nums = np.linspace(x1-1, x+1, 10)
    ax.plot(x, f(x), 'r')
    ax.plot(x_nums,line(x_nums, x1), 'b--', linewidth = 0.3)
    ax.scatter(x1, y1, color = 'b', s = 20)
    ax.plot(np.linspace(ap_y, ap_y, 10),line(x_nums, x1), 'g--', linewidth = 0.3)
```

```python
interactive(plot, m=(-30,30, 1))
```

## Task 2: Newton's Method
<p>Implement a function that takes a differentiable function and returns any root of that function.</p>

```python
def plot(func_x: str, root: float):
    
    font = {'weight': 'bold', 'size': 9}
    matplotlib.rc('font', **font)
    fig, ax = plt.subplots()
    ax.spines["left"].set_position(("data", 0))
    ax.spines["bottom"].set_position(("data", 0))
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.plot(1, 0, ">k", transform = ax.get_yaxis_transform(), clip_on = False)
    ax.plot(0, 1, "^k", transform = ax.get_xaxis_transform(), clip_on = False)
    
    x_nums = np.linspace(-3, 3, 101)
    x = sp.Symbol('x')
    func_x = sp.sympify(func_x)
    
    y_nums = [func_x.subs(x, num) for num in x_nums]
    ax.plot(x_nums, y_nums, 'r', linewidth = 2)
    ax.scatter(root, 0, color = 'b', s = 30)
    
```

```python
x = sp.Symbol('x')

def get_root(func_x):
    func_x = sp.sympify(func_x)
    der_fx = sp.diff(func_x)
    root = 0.1
    for n in range(32):
        if abs(func_x.subs(x, root)) < 0.0000001:
            return root
        if der_fx.subs(x, root) != 0:
            root = root - func_x.subs(x, root)/der_fx.subs(x, root)

user_function = input("enter your function like: x**2: ")



print("root: ", get_root(user_function))
plot(user_function, get_root(user_function))
```
