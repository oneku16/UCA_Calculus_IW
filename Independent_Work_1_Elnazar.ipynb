## Limit of a Function

<p>In this independent work, you are going to analyze one of the indeterminate form 
$1^ \infty $ </p>

<p>For the simplicity, we will be looking at the limit as $x$ approaches 0 from left; that is, $x \to 0 ^ {-} $


```python
%matplotlib notebook
%matplotlib inline
import numpy as np
import sympy as sp
import matplotlib
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from IPython.display import display
from pylab import *
```

```python
def graph_plot(x_axis, y_axis, color: str):
    font = {'weight' : 'bold',
    'size'   : 18}
    matplotlib.rc('font', **font)
    desk = plt.figure(figsize = (30, 20))
    graph = desk.add_subplot(111, aspect='equal')
    graph.set_xbound(-5,5)
    graph.set_ybound(-10,10)
    sub1 = subplot(111)
    autoAxis = sub1.axis()
    plt.plot(x_axis, y_axis, color, linewidth = 1)
```

## Task 1:
<p><ul>
<li> Find a function $f(x)$ (preferably rational function) such that <br> $\lim_{x \to 0^{-}} f(x) = 0$ </li>
<li> Show that the limit is 0 (by making a table of values) </li>
<li> Draw the graph of the function in the interval $[-3,3]$ </li></ul></p>

```python
x = sp.Symbol('x')
```

```python
vals = np.array([-1, -0.1, -0.01, -0.001, -0.0001, -0.00001, 0])
```

```python
func_f = sp.lambdify(x, (x**3 - 2*x)/(2*(x**2-5)))
```

```python
x_nums = np.linspace(-3, 3, 11)
```

```python
for val in vals:
    print(func_f(val))
```

```python
list_f_x = np.asarray(list_f_x)
list_f_x
```

```python
graph_plot(x_nums, func_f(x_nums), "or")
```

## Task 2:

```python
graph_plot(x_nums, func_g(x_nums), "ob")
```

## Task 3:
<p>Now evaluate the limit (by making table of values and drawing the graph)<br> $\lim_{x \to 0^{-}} {(1+f(x))}^{g(x)}$</p>

```python
func_fg = sp.lambdify(x, np.power(1 + (x**3 - 2*x)/(2*(x**2-5)),(1/x**2)))
```

```python
for val in vals:
    print(func_fg(val))
```

```python
graph_plot(x_nums, func_fg(x_nums), "g")
```
